# Phân tích hiệu năng Cờ Caro AI (Minimax vs Alpha-Beta)

Notebook này phân tích dữ liệu benchmark được sinh ra từ `runner.py`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu
df = pd.read_csv('logs/benchmark_results.csv')
df.head()
```

## 1. So sánh Số lượng Node phải duyệt

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Depth', y='Nodes', hue='Algorithm', errorbar=None)
plt.title('Số lượng Node phải duyệt theo Độ sâu (Càng thấp càng tốt)')
plt.yscale('log') # Thang đo logarit vì chênh lệch quá lớn
plt.ylabel('Số Nodes (Log scale)')
plt.show()
```

## 2. So sánh Thời gian chạy

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Depth', y='Time_ms', hue='Algorithm', marker='o')
plt.title('Thời gian tính toán theo Độ sâu (Càng thấp càng tốt)')
plt.ylabel('Thời gian (ms)')
plt.show()
```
